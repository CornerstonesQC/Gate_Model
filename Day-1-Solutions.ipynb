***Exercise 1.*** Recall that our normalization condition gives

$$
|\alpha|^2 + |\beta|^2 = (\mbox{Re} \alpha)^2 + (\mbox{Im} \alpha)^2 + (\mbox{Re} \beta)^2 + (\mbox{Im} \beta)^2 = 1.
$$

If we use phase ambiguity to make $\alpha$ real, then $\alpha = \mbox{Re} \alpha$ and $\mbox{Im} \alpha=0$, so

$$
\alpha^2 + (\mbox{Re} \beta)^2 + (\mbox{Im} \beta)^2 = 1,
$$

describing a point on the sphere. See below for some highly non-essential reading about this sphere.

*Spherical subtleties.* There is a technicality we've deliberately swept under the rug. The constraint that $\alpha \geq 0$ means we have the northern *hemisphere* rather than a full sphere. In order to get a full sphere, with $\alpha \in [-1, 1]$, we could replace $\theta/2 \mapsto \theta$. The issue is that global phase ambiguity identifies the antipodal points on the big sphere, $(\alpha, \beta) \sim (-\alpha, -\beta)$, so we would cover the space of states *twice*. On the Bloch sphere, we want one point per state, so we throw away the southern hemisphere, and contract the equator $\alpha = 0, \beta = e^{i\phi}$ to a single point, once again using phase ambiguity on $\beta$. Geometrically, we pinch the big sphere at the equator, splitting it into two spheres, and just keep one of them!

*Hopf for the best.* Under the mathematical hood, something remarkable is going on. We started with a 3-sphere in $\mathbb{C}^2$, defined by $|\alpha|^2 + |\beta|^2 = 1$. The phase ambiguity $(\alpha,\beta) \sim e^{i\phi}(\alpha,\beta)$ pinches circles, aka 1-spheres, into points, and the result is the Bloch 2-sphere. Putting it differently, we can construct the 3-sphere by pasting 1-spheres in a "twisted" fashion onto the surface of a 2-sphere, a gorgeous result called the [*Hopf fibration*](https://en.wikipedia.org/wiki/Hopf_fibration). You can find many pretty visualizations online, for instance, [this hypnotic video](https://www.youtube.com/watch?v=AKotMPGFJYk) by Niles Johnson.

***Exercise 3.*** (a) We know that $Z|0\rangle = |0\rangle$ and $Z|1\rangle = -|1\rangle$. For a single qubit, the phase is unimportant, so we can get rid of $-1 = e^{i\pi}$ with impunity. The result is that $Z$ leaves both alone! But for $|\pm\rangle$, we have

$$
Z|\pm\rangle = \frac{1}{\sqrt{2}}(Z|0\rangle \pm Z|1\rangle ) = \frac{1}{\sqrt{z}}(|0\rangle \mp |1\rangle ) = |\mp\rangle.
$$

In other words, $Z$ *swaps* $|+\rangle$ and $|-\rangle$.

(b) If it is rotation of the sphere, it keeps $|0\rangle$ and $|1\rangle$ fixed, suggesting that the axis is the north-south pole. If it swaps two opposite points, it must be rotating by an angle $\pi$ around this axis.

(c) Let's play the same guessing game with $X$. First, we note that $X|0\rangle = |1\rangle$, $X|1\rangle = |0\rangle$, and $X|\pm\rangle = \pm |\pm\rangle$. If this is a rotation, it suggests that the sphere rotates by $\pi$ around the axis between $|\pm\rangle$.

---
***Bonus 1.*** (a) In terms of Bloch coordinates, rotate by $\pi$ around the north-south pole simply maps $\phi \mapsto \phi + \pi$. By acting on $|\psi(\theta,\phi)\rangle$ with $Z$, we can check this is precisely what happens for an arbitrary state:

\begin{align*}
Z|\psi(\theta,\phi)\rangle & = \cos \left(\frac{\theta}{2}\right)Z|0\rangle + \sin \left(\frac{\theta}{2}\right) e^{i\phi}Z|1\rangle \\
&= \cos \left(\frac{\theta}{2}\right)|0\rangle + \sin \left(\frac{\theta}{2}\right) e^{i\phi}(-1)|1\rangle \\
&= \cos \left(\frac{\theta}{2}\right)|0\rangle + \sin \left(\frac{\theta}{2}\right) e^{i(\phi+\pi)}|1\rangle = |\psi(\theta,\phi+\pi)\rangle.
\end{align*}

(b)  Using a double angle formula, we have

$$
\langle \psi(\theta,\phi)|Z|\psi(\theta,\phi)\rangle = \cos^2\left(\frac{\theta}{2}\right) - \sin^2\left(\frac{\theta}{2}\right) = \cos \theta.
$$

(c)

We'll start with $X$:

\begin{align}
\langle X \rangle & = \left[\cos \left(\frac{\theta}{2}\right)\langle 0|+ e^{-i\phi}\sin \left(\frac{\theta}{2}\right) \langle 1|\right]X\left[\cos \left(\frac{\theta}{2}\right)|0\rangle + e^{i\phi}\sin \left(\frac{\theta}{2}\right) |1\rangle\right] \\
& = \left[\cos \left(\frac{\theta}{2}\right)\langle 0|+ e^{-i\phi}\sin \left(\frac{\theta}{2}\right) \langle 1|\right]\left[\cos \left(\frac{\theta}{2}\right)|1\rangle + e^{i\phi}\sin \left(\frac{\theta}{2}\right) |0\rangle\right] \\
& = \frac{1}{2}(e^{i\phi} + e^{-i\phi})\cdot 2\cos \left(\frac{\theta}{2}\right)\sin \left(\frac{\theta}{2}\right) \\
& = \cos\phi \sin\theta.
\end{align}

The calculation for $Y$ is almost identical, but we end up with

$$
\langle Y\rangle = \frac{1}{2}(-ie^{i\phi} + i e^{-i\phi}) \cdot 2\cos \left(\frac{\theta}{2}\right)\sin \left(\frac{\theta}{2}\right) = \sin\phi\sin\theta.
$$

---
***Bonus 2.*** (a) Proving this rigorously takes a bit of work, so here, we will instead rely on geometric intuition. To rotate a sphere, you need to choose an axis to rotate around, and some angle to rotate through. The axis joins two antipodal points of the sphere which are fixed, and moves everything else (with exceptions we discuss in (c)). These two antipodal points are precisely the eigenvectors!

(b) Here, we use the basic fact about linear algebra that unitary matrices have orthogonal eigenvectors. Since the two antipodal points correspond to distinct eigenvectors for some unitary matrix (acting as a rotation on the sphere), they are orthogonal. If two states *are* orthogonal, we can always use them to define a unitary matrix, for which they will be eigenvectors and hence antipodal points. So it goes both ways!

(c) If an operator is some multiple of the identity, then every point on the sphere gets mapped to itself. (There may be a global phase factor, but this is irrelevant due to phase ambiguity.) The only way this is possible, given that we are rigidly rotating the sphere, is if the angle of rotation is some integer multiple of $2\pi$. It follows that if $U^k \propto I$, then the angle of rotation $k\vartheta$ must be some integer multiple of $2\pi$. A unitary need not have this property, e.g. it rotates by an irrational number multiple of $2\pi$.

---
***Bonus 3.*** (a) Let's use Bonus 2(a). Since $X^2 = I$, we know that $2\vartheta$ is a multiple of $2\pi$. Since $X$ itself is nontrivial, we can take $\vartheta = \pi$. We can find the axis by observering that the eigenvectors are $|\pm\rangle$, since for instance

$$
X|+\rangle = \frac{1}{\sqrt{2}}(X|0\rangle + X|1\rangle) = \frac{1}{\sqrt{2}}(|1\rangle + |0\rangle) = |+\rangle.
$$

So we can take $\hat{\mathbf{n}}$ to lie at $|+\rangle = |\psi(\pi, 0)\rangle$.

(b) The same reasoning we used for $X$ shows that $H$ must rotate through an angle $\pi$.

(c) Here, we can use some algebraic trickery. Since $|+\rangle = H|0\rangle$, and $H^2 = I$, we have

$$
H \cdot \frac{1}{\sqrt{2}}(|0\rangle + |+\rangle) = H \cdot \frac{1}{\sqrt{2}}(I + H)|0\rangle = \frac{1}{\sqrt{2}}(H + H^2)|0\rangle = \frac{1}{\sqrt{2}}(H + I)|0\rangle = \frac{1}{\sqrt{2}}(|0\rangle + |+\rangle).
$$

To find the antipodal points, we must determine the antipodal point to $\frac{1}{\sqrt{2}}(|0\rangle + |+\rangle)$. We can use Exercise 4(b), namely the fact that points are antipodal if and only if the corresponding states are orthogonal. There are many ways to skin this bra-ket, but we can use our $H^2 = I$ trick once more:

\begin{align}
\frac{1}{2}(\langle 0| + \langle +|)(|1\rangle - |-\rangle) =& \frac{1}{2}\langle 0|(H+I)(I-H) |-\rangle = \frac{1}{2}\langle 0|(I-H^2) |1\rangle = 0.
\end{align}

So they are antipodal. Of course, it is simpler to just argue from orthogonality, but the point of the Bloch sphere is to geometrize as much as possible!

(d) We could calculate this in the usual coordinates, but it's simpler to set up spherical coordinates around the eigenvectors $|H_\pm\rangle$ of $H$, and expand a state as

$$
|\psi\rangle = \cos\left(\frac{\theta}{2}\right) |H_+\rangle + e^{i\phi}\sin\left(\frac{\theta}{2}\right)\beta |H_-\rangle.
$$

Then the same calculation as Exercise 3(d) gives

$$
\langle H\rangle = \cos^2\left(\frac{\theta}{2}\right) - \sin^2\left(\frac{\theta}{2}\right) = \cos\theta.
$$

Once again, this is a projection onto the axis running between the antipodal eigenvectors. This observation generalizes to any $U^2 = I$, simply because the eigenvalues are $\pm 1$. (This follows because the eigenvalues must satisfy $\lambda^2 = 1$, and if both eigenvalues are the same, $U$ is a trivial multiple of the identity.) Setting up spherical coordinates around the axis between $|U_\pm\rangle$, the same calculation goes through, and in general, $\langle U\rangle$ is the projection of the Bloch vector onto the axis between it eigenvectors.

## Quantum Circuits in Pennylane

In this we will introduce quantum computation and PennyLane. If you’ve never evaluated a quantum circuit before, this is a great place to start. Even if you are familiar with quantum computation, this challenge will introduce you to the syntax used for PennyLane circuits.

### Define a device

A device object in Pennylane provides the possibility of evaluating a quantum function - either with a simulator or an actual quantum hardware as backend. The code below instantiates a device using the `default.qubit` backend, which is a built-in simulator that does not require external dependencies and runs on your very own machine!

````python
dev_1qubit = qml.device('default.qubit', wires=1, shots=1000)
````

Further parameters to specify when setting up a device are:

-  **Number of wires**: Pennylane's generic term for qubits. Gates and measurements always act on these wires. Each qubit are always initiated in the state $\ket{0}$ of Pauli $Z=+1$ when using the `default.qubit` simualtor.
- **Number of shots**: computations on real QPUs are stochastic processes, following probability distributions. The computations need to be repeated **shots** times to acquire a result with statistical significance. Simulators are able to extract information of quantum states as if performing analytical calculations with $\bra{\,\,}$ and $\ket{\,\,}$, hence this parameter does not matter here.
- See [Pennylane Docs](https://pennylane.readthedocs.io/en/stable/code/api/pennylane.device.html) for further details.

### Defining a Quantum Circuit - The QNode decorator

The device is then used to construct so called `QNodes`, which are representations of quantum nodes in a computational graph. In other words, it contains a function representing our quantum circuit and the device we defined. Pennylane uses python decorators (`@decorator`) to wrap a quantum function.

````python
# Create a quantum circuit and qnode
@qml.qnode(dev_1qubit)
def circuit():
    #measure the probability of finding our 
    #single qubit in the computational basis states 0,1.
    return qml.probs(0)
````

Here, our quantum function is `circuit()` and does not depend on any parameter. The return value is an array of probabilities $|\langle i | \psi \rangle |^2$ for measuring the computational basis state $|i\rangle$ given the current state $|\psi \rangle$.

If we execute this basic circuit, we find a probability of $1$ for our single qubit to be in the computational state $0$, which is what it was initiated to.

````python
print("Output array [p(|0>),p(|1>)]: {0}".format(circuit()))
````

## Exercise 2: Basis Transformations

Qbits, the operational units of quantum computers are vectors in 2-dimensional Hilbert spaces. Hilbert spaces are spanned by basis vectors; multiple bases exist for a given Hilbert space. For quantum computation it is important to learn how to operate and change bases.


The choice of basis is determined according to the 3 Pauli matrices $X$, $Y$, $Z$. Conventionally we will call the `computational basis` that which makes $Z$ diagonal for every qubit: 

$$
Z=\begin{pmatrix}
1 & 0\\
0 & -1
\end{pmatrix}
;\ \ket{0}= \begin{pmatrix}
1 \\
0 
\end{pmatrix}
;\ \ket{1}= \begin{pmatrix}
0 \\
1 
\end{pmatrix}
$$

This is just the basis we have been using for our calculations above.

The PauliZ matrix is also known as a *phase flip* operator (as it adds a phase $\pi$ to $\ket{1}$ with respect to $\ket{0}$). In this basis, the Pauli $X$ operator is a *bit flip* operator and reads 

$$ 
X=\begin{pmatrix}
0 & 1\\
1 & 0
\end{pmatrix} 
$$

Naturally, $X$ can be diagonalized to a basis $\ket{\pm}$ which, in terms of the computational basis, reads

$$
\ket{\pm}=\frac{\ket{0} \pm \ket{1}}{\sqrt{2}}
$$

### 4.2 Bell State
Some interesting, maximally entangled 2-qubit states are **Bell states** 

$$ \begin{aligned} \ket{\Phi_+} &= \frac{1}{\sqrt{2}} \left( \ket{00} + \ket{11} \right)\\ \ket{\Phi_-}&= \frac{1}{\sqrt{2}} \left( \ket{00} - \ket{11} \right)\\ \ket{\Psi_+} &= \frac{1}{\sqrt{2}} \left( \ket{01} + \ket{10}\right)\\ \ket{\Psi_-} &= \frac{1}{\sqrt{2}} \left( \ket{01} - \ket{10}\right) \,. \end{aligned} $$

If the system is in state $\ket{\Phi^+}$ or $\ket{\Phi^-}$ then if one qubit is measured as $\ket{0}$ then we immediately know the other qubit is also in state $\ket{0}$ (and similarly with $\ket{1}$). If the system is in state $\ket{\Psi^\pm}$ then if one qubit is measured we immediately know the other qubit is in the opposite state.

Show that a Hadamard gate followed by a CNOT applied to $\ket{00}$ produces the Bell state $\ket{\Phi^+}$.


### Solution

H_0 \ket{00} &= \frac{1}{\sqrt{2}}\left(\ket{00}+\ket{10}\right) \\

\end{aligned}
$$

### 4.3 Expectation Value of a Multi-qubit state
Measure the expectation value of an entangled state wrt a tensor-product observable.

• Instantiate a new device with 2 qubits.

• Create the **Bell-state**:
$$|\Phi^+>=\frac{1}{\sqrt{2}}(|00>+|11>)$$

• Rotate the first qubit around the y-axis by an angle $\theta=\frac{\pi}{4}$

$$|\theta>=R^0_y(\theta)|\Phi^+>=e^{-i\theta(\sigma_y\otimes I)/2}|\Phi^+>$$

• Measure the expectation value of the state wrt to the tensor product observable Pauli-Z(0)@Pauli_Z(1)
$$|<\theta|\sigma_z^0\otimes\sigma_z^1|\theta>|^2$$

````python
#dev_2qubit = qml.device(...)

#SOLUTION
dev_2qubit = qml.device('default.qubit', wires=2)
````

````python
def entangledstate(angle):
    expectation_value = 0.0

    # Create a quantum circuit and qnode
    @qml.qnode(dev_2qubit)
    def circuit(par):
        #1.) create |+> from first qubit |0>
        #2.) connect first and second qubit with a CNOT gate
        #3.) apply a y-rotation gate on the first qubit
        #4.) measure the expectation value. of the PauliZ-observable Pauli-Z(0)@Pauli_Z(1)
#         return
    
        #solution
        qml.Hadamard(0) #create psi=1/sqrt(2)(|0>+|1>) on qubit 0
        qml.CNOT(wires=[0,1]) #qubit 0 is control of qubit 1
        qml.RY(par, wires=0)# a single-wire parameterized gate
        return qml.expval(qml.PauliZ(0) @ qml.PauliZ(1))


    expectation_value = circuit(angle)

    print(circuit.draw())
    return expectation_value
````

````python
#Execute this after adjusting the circuit
angle=np.pi/4
expval=entangledstate(angle)
print("Expectation value: {0:.3f}".format(expval))
````
